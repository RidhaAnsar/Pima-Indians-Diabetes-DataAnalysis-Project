```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
diabetes=pd.read_csv("drive/MyDrive/diabetes/diabetes.csv")
diabetes.head()
```

```python
diabetes.tail()
```

```python
# create a histogram of age
diabetes["Age"].plot.hist(bins=20);
```

```python
diabetes.plot.hist(figsize=(10,30),subplots=True);
```

```python
# data analysis of patients of age above 50
over_50=diabetes[diabetes["Age"]>50];
over_50.head()
```

```python
over_50.plot(kind='scatter',
             x='Age',
             y='BloodPressure',
             c='Outcome')
```

```python
over_50.plot(kind='scatter',
             x='Age',
             y='BloodPressure',
             c='Outcome',
             figsize=(10,6),
             subplots=True);
```

```python
# object oriented plot
fig, ax=plt.subplots(figsize=(10,6));
scatter=ax.scatter(x=over_50["Age"],
                   y=over_50["BloodPressure"],
                   c=over_50["Outcome"])

ax.set(title="Diabetes and bloodpressure levels",
       xlabel="Age",
       ylabel="BloodPressure")

#adding a legend
ax.legend(*scatter.legend_elements(),
          title="Outcome")
ax.axhline(over_50['BloodPressure'].mean())
```
